```python
import random
import json

tones = {
    "warm": ["#FF5733", "#FF8D1A", "#FFC300", "#C70039", "#A04000"],
    "cool": ["#3498DB", "#2ECC71", "#1ABC9C", "#5DADE2", "#48C9B0"],
    "neutral": ["#FFFFFF", "#D5D8DC", "#AAB7B8", "#808B96", "#2C3E50"],
    "earthy": ["#6B8E23", "#8B5E3C", "#556B2F", "#A0522D", "#8FBC8F"],
    "pastel": ["#FFD1DC", "#C1E1C1", "#D6EAF8", "#F9E79F", "#E8DAEF"]
}

styles = ["cozy", "modern", "luxury", "vibrant", "minimal", "nature"]

style_tone_map = {
    "cozy": ["warm", "pastel", "earthy"],
    "modern": ["neutral", "cool"],
    "luxury": ["neutral", "warm"],
    "vibrant": ["warm", "cool"],
    "minimal": ["neutral"],
    "nature": ["earthy", "cool"]
}

def adjust_color(hex_color):
    # Slight variation for realism
    hex_color = hex_color.lstrip("#")
    r = int(hex_color[0:2], 16)
    g = int(hex_color[2:4], 16)
    b = int(hex_color[4:6], 16)

    r = min(255, max(0, r + random.randint(-20, 20)))
    g = min(255, max(0, g + random.randint(-20, 20)))
    b = min(255, max(0, b + random.randint(-20, 20)))

    return f"#{r:02X}{g:02X}{b:02X}"

def generate_palette():
    style = random.choice(styles)
    tone = random.choice(style_tone_map[style])

    base_colors = tones[tone]

    # pick 3 base colors
    colors = random.sample(base_colors, 3)

    # add 2 realistic variations
    colors += [adjust_color(random.choice(base_colors)) for _ in range(2)]

    return {
        "style": style,
        "tone": tone,
        "colors": colors
    }

# Generate 2000 palettes
dataset = []

for i in range(2000):
    palette = generate_palette()
    palette["id"] = i
    dataset.append(palette)

with open("palettes_realistic_2000.json", "w") as f:
    json.dump(dataset, f, indent=4)

print("🔥 Realistic 2000 palettes generated!")
```
